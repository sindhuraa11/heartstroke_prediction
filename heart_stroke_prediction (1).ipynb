```python
#convert categorical data into numerical format
from sklearn.preprocessing import LabelEncoder
```

```python

encoder=LabelEncoder()
```

```python
columns=['gender','ever_married','work_type','Residence_type','smoking_status']
```

```python
df.head()
```

```python
df.gender=encoder.fit_transform(df.gender)
df.head()
```

```python
columns
```

```python
df.ever_married=encoder.fit_transform(df.ever_married)
```

```python
for column in columns:
    df[column]=encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
grouped_data = df.groupby('gender')['stroke'].sum()

# Create the bar plot
plt.bar(grouped_data.index, grouped_data.values)

# Customize the plot
plt.xlabel('Gender')
plt.ylabel('Number of Strokes')
plt.title('Stroke Count by Gender')
```

```python
grouped_data = df.groupby('age')['stroke'].sum()

# Create the bar plot
plt.bar(grouped_data.index, grouped_data.values)

# Customize the plot
plt.xlabel('age')
plt.ylabel(' heart Stroke')
plt.title('Stroke Count by age')
```
